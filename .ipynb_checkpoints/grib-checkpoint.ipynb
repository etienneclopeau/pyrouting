```python
grbs = pygrib.open('grib/NorthEurope.wind.7days.grb')
```

```python
for grb in grbs[:10]:
    print(grb)
```

```python
#print(grb.keys())
```

```python
grbs.rewind()
for grb in grbs[:10]:
    print(grb.validityDate,grb.validityTime,grb.name)
```

```python
grbU, grbV = grbs[1:3]
#grb.latlons()
U, lats,lons = grbU.data()
V = grbV.data()[0]
print(U,V,lats,lons)
projd = grbU.projparams
print(projd)
```

```python
latmin = lats.min()
latmax = lats.max()
lonmin = lons.min()
lonmax = lons.max()
```
